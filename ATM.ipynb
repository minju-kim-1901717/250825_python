```python
# 전체 ATM 코드

# 입금 = deposit_amount
# 출금 = withdraw_amount

# 잔액 = balance
balance = 0
# 거래내역을 저장할 리스트
receipts = []

print(f"현재 잔액 : {balance}원")

while True:
    print("----*------------*----")
    print("어떤 업무를 진행하실건가요")
    user_choice = input("| 입금 | 출금 | 잔액확인 | 거래내역 | 종료 |")

    if user_choice == "입금":
        # 입금 금액을 사용자로부터 입력받음
        deposit_amount = int(input("입금 금액을 입력해주세요: "))

        # 잔액 + 입금 금액
        balance += deposit_amount

        # receipt에 입금 내역 저장
        receipts.append(("입금", deposit_amount, balance))

        # 입금 금액 및 현재 잔액 출력
        print(f"입금 금액 : {deposit_amount}원")
        print(f"현재 잔액은 {balance}원 입니다")

    elif user_choice == "출금":
        # 출금 금액을 사용자로부터 입력받음
        withdraw_amount = int(input("출금 금액을 입력해주세요: "))

        # 잔액(balance)에서 출금
        # 입력금액 > 잔액 이라면 입력금액 = 잔액으로 변경
        if withdraw_amount > balance:
            withdraw_amount = balance

        # 잔액 - 출금 금액
        balance -= withdraw_amount

        # receipts 출금 거래 정보 튜플 추가
        receipts.append(("출금", withdraw_amount, balance))

        # 현재 잔액 출력
        print(f"출금 금액 : {withdraw_amount}원")
        print(f"현재 잔액은 {balance}원 입니다")

    # 잔액 확인
    elif user_choice == "잔액확인":
        print(f"현재 잔액 : {balance}원")

    # 거래 내역 영수증 확인
    elif user_choice == "거래내역":
        for receipt in receipts:
            print(f"{receipt[0]} : {receipt[1]}원")
        print(f"현재 잔액 : {balance}원")

    # 거래 종료
    elif user_choice == "종료":
        print("ATM을 종료합니다.")
        break

    # 잘못된 입력값 입력시
    else:
        print("잘못된 입력입니다.")
```
