__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
ma = max(nums)
mi = min(nums)
avg = sum(nums)/len(nums)

print(f"최댓값 : {ma}")
print(f"최솟값 : {mi}")
print(f"평균 : {avg}")
````

__[문제7]__

아래 순서대로 주어진 수강생 이름과 등급을 하나의 딕셔너리로 만드는 코드를 작성해주세요


names : 나이썬, 김자바, 김씨, 박씨샵
