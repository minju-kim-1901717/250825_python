```python
# 여기에 코드 작성

# 1. 사용자에게 숫자로 입력값을 받는다
# 2. 입력받은 값으로 반복문을 돌린다
# 3. 반복문 안에서 랜덤으로 6개의 숫자를 생성한다

import random

def lotto_numbers():
    # 로또 번호 6개 생성 (1~45, 중복 없음)
    numbers = random.sample(range(1, 46), 6)
    numbers.sort()
    return numbers

print("로또 번호 생성기")


while True:
    try:
        print("🍀" * 30)
        count = int(input("몇 개의 번호를 생성할까요? (0은 종료): "))

        if count == 0:
            print("프로그램을 종료합니다.")
            break

        if count < 0:
            print("0 이상의 숫자를 입력해주세요")
            continue

        print(f"\n🍀{count}개의 로또 번호🍀")
        print("-" * 25)

        for i in range(count):
            nums = lotto_numbers()  # ✅ 함수 호출
            print(f"{i+1}번째: {', '.join(map(str, nums))}")

        print(f"\n총 {count}개의 번호가 생성되었습니다!")
        print("🍀행운을 빕니다🍀\n")

    except ValueError:
        print("숫자를 입력해주세요!")
```
