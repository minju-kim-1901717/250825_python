### 프로젝트: 간단한 ATM 관리 시스템
<br>

__[문제 1: 잔액 확인]__

현재 잔액은 1000원입니다. 현재 잔액을 출력하세요.

(변수명 예시 : 잔액 - balance)

```python
# 잔액 - balance
balance = 1000
print(f"현재 잔액은 {balance}원 입니다")
```

__[문제 2: 입금]__

가지고 있는 돈을 입금합니다.

입금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 입금 거래를 영수증 리스트에 기록하세요.

- 입금 금액은 input()을 이용해 받습니다.
- 영수증은 list()로 선언합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 영수증 리스트 - receipts, 입금 금액 - deposit_amount)


```python
# 입금 금액을 사용자로부터 입력받기
deposit_amount = int(input("입금 금액을 입력해주세요: "))

# 잔액(balance) + 입금 금액(deposit_amount)
balance += deposit_amount

# 영수증 리스트 - receipts, 거래 정보를 튜플 형태로 저장
receipts = []
receipts.append(("입금", deposit_amount, balance))

# 잔액 출력
print(f"현재 잔액은 {balance}원 입니다")
```

__[문제 3: 출금]__

사용자가 돈을 출금합니다. 출금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 출금 거래를 영수증 리스트에 기록하세요.

- 출금 금액은 input()을 이용해 받습니다.
- 현재 잔액보다 출금 요청 금액이 많을 경우 현재 잔액 만큼만 출금되도록 합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 출금금액 - withdraw_amount)

```python
# 출금 금액 입력받기
withdraw_amount = int(input("출금 금액을 입력해주세요: "))

# 잔액(balance)에서 출금
# 입력받은 금액 >= 잔액이면 출금 금액 = 잔액으로
if withdraw_amount > balance:
    withdraw_amount = balance

# 잔액(balance)에서 출금 금액을 빼고 갱신
balance -= withdraw_amount

# receipts 리스트에 출금 거래 정보를 튜플로 추가
receipts.append(("출금", withdraw_amount, balance))

# 현재 잔액을 출력
print(f"현재 잔액은 {balance}원 입니다")
```

__[문제 4: 모든 거래 내역 확인]__

모든 거래 내역(영수증)을 요청받았을 때 출력하세요.
- 지금까지 영수증에 추가된 정보를 출력합니다.
- 출력 형태는 자유입니다.

```python
# receipts 리스트에 저장된 모든 거래를 하나씩 출력해 보세요.
# for문을 이용해서 각 거래 정보를 반복하며 출력하면 됩니다.
# 출력 형식은 자유롭게 구성해도 좋아요.


print(f"현재 잔액 : {balance}원")

for receipt in receipts:
        print(f"{receipt[0]} : {receipt[1]}원")
```

```python
# 전체 ATM 코드

# 입금 = deposit_amount
# 출금 = withdraw_amount

# 잔액 = balance
balance = 0
# 거래내역을 저장할 리스트
receipts = []

print(f"현재 잔액 : {balance}원")

while True:
    print("----*------------*----")
    print("어떤 업무를 진행하실건가요")
    user_choice = input("| 입금 | 출금 | 잔액확인 | 거래내역 | 종료 |")

    if user_choice == "입금":
        # 입금 금액을 사용자로부터 입력받음
        deposit_amount = int(input("입금 금액을 입력해주세요: "))

        # 잔액 + 입금 금액
        balance += deposit_amount

        # receipt에 입금 내역 저장
        receipts.append(("입금", deposit_amount, balance))

        # 입금 금액 및 현재 잔액 출력
        print(f"입금 금액 : {deposit_amount}원")
        print(f"현재 잔액은 {balance}원 입니다")

    elif user_choice == "출금":
        # 출금 금액을 사용자로부터 입력받음
        withdraw_amount = int(input("출금 금액을 입력해주세요: "))

        # 잔액(balance)에서 출금
        # 입력금액 > 잔액 이라면 입력금액 = 잔액으로 변경
        if withdraw_amount > balance:
            withdraw_amount = balance

        # 잔액 - 출금 금액
        balance -= withdraw_amount

        # receipts 출금 거래 정보 튜플 추가
        receipts.append(("출금", withdraw_amount, balance))

        # 현재 잔액 출력
        print(f"출금 금액 : {withdraw_amount}원")
        print(f"현재 잔액은 {balance}원 입니다")

    # 잔액 확인
    elif user_choice == "잔액확인":
        print(f"현재 잔액 : {balance}원")

    # 거래 내역 영수증 확인
    elif user_choice == "거래내역":
        for receipt in receipts:
            print(f"{receipt[0]} : {receipt[1]}원")
            print(f"현재 잔액 : {balance}원")

    # 거래 종료
    elif user_choice == "종료":
        print("ATM을 종료합니다.")
        break

    # 잘못된 입력값 입력시
    else:
        print("잘못된 입력입니다.")
```
